```python
import pandas as pd

df = pd.read_csv("data/ihdp.csv")
df.head()
```

```python
df = df.rename(columns=str.lower)

features = [c for c in df.columns if c not in ["treatment", "y_factual", "y_cfactual"]]
X = df[features]
t = df["treatment"]
y = df["y_factual"]
```

```python
from dowhy import CausalModel

model = CausalModel(
    data=df,
    treatment="treatment",
    outcome="y_factual",
    common_causes=features
)

identified = model.identify_effect()
estimate = model.estimate_effect(identified, method_name="backdoor.propensity_score_matching")
estimate.value
```
